# Creating Fake Dataset for StartOut Project

**Synopsis of Documentation and Videos:**

- **Mean:**  average experience for the LGBT founder in these cities
    * be wary of small sample size
- **Size:**  absolute scale of average experience for the LGBT entrepreneur
- **Color:**  achievement gap - difference between the mean achievement and what we truly believe is possible

- **Actual (counted):** out entrepreneurs)
- **Actual Estimated:** include not 
- **Achievable:** Prediction based on best in class comparison

On City View:
- Comparing Total vs. Mean
- Comparing LGBT vs. Everyone Else (on/off)
- Comparing Actual vs. Actual (est.)
    - indication of out vs. non-out

- **High-growth entrepreneurs:**
    - entrepreneurs who were co-founders or employees prior to their company meetings at least the criteria below:


- **High-growth companies:** 
    * received any amount of *Venture Capital Funding*, or 
    * received at least $1M in *Angel Funding*, or
    * generated at least one patent **and** has created jobs beyond the founding team, or
    * had an IPO or been acquired by another company.

- **Metropolitan area:**
    - minimum of 50 growth entrepreneurs in total (incl. LGBTQ+)

- **Results:**
1. 62,074 High-growth companies
2. 124,756 high-growth entrepreneurs
    - identified 6703 LGBT and 92096 straight or unknown
    - about 7%
3. 99 Metro Areas

```python
def random_genders(size, p=None):
    if not p:
        p = (0.49, 0.49, 0.01, 0.01)
    gender = ("M", "F", "NB", "O")
    np.random.seed(4321)
    return np.random.choice(gender, size=size, p=p)

founders['Gender'] = list(random_genders(size_entrepreneurs))
```

According to a Gallup poll released Wednesday, 5.6% of United States adults identify as LGBT.  The poll found that more than half of LGBT adults (54.6%) identify as bisexual, about a quarter (24.5%) as gay, 11.7% as lesbian and 11.3% as transgender. An additional 3.3% used a different non-heterosexual term to describe their sexual orientation, such as queer or same-gender-loving.
```python
def random_sexuality(size, p=None):
    if not p:
        p = (0.03, 0.014, 0.944, 0.007, 0.005)
    sexuality = ('B', 'G', 'H', 'L', 'O')
    np.random.seed(4321)
    return np.random.choice(sexuality, size=size, p=p)

founders['Sexuality'] = random_sexuality(size_entrepreneurs)
```

```python
def random_race(size, p=None):
    if not p:
        p = (0.601, 0.185, 0.134, 0.059, 0.013, 0.008)
    race = ('White', 'Hispanic', 'Black', 'Asian', 'AmIndian', 'Other')
    np.random.seed(4321)
    return np.random.choice(race, size=size, p=p)

founders['Race'] = random_race(size_entrepreneurs)
```

```python
def random_transgender(size, p=None):
    if not p:
        p = (0.113, 0.887)
    transgender = ("Y", "N")
    np.random.seed(4321)
    return np.random.choice(transgender, size=size, p=p)

founders['Transgender'] = random_transgender(size_entrepreneurs)
```
